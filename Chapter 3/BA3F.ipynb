## BA3F

**Eulerian Cycle Problem**

Find an Eulerian cycle in a graph.

Given: An Eulerian directed graph, in the form of an adjacency list.

Return: An Eulerian cycle in this graph.

Link: https://rosalind.info/problems/ba3f/

```python
import random
```

```python
def input_to_graph(adjacency_list):
    dictionary = {}
    for edge in adjacency_list:
        first, second = edge.split(" -> ")
        second = second.split(",")
        dictionary[first] = second
    return dictionary
```

```python
def find_eulerian_cycle(graph,node):
    cycle = [node]
    while graph[node]:
        cycle = cycle[:1] + find_eulerian_cycle(graph,graph[node].pop()) + cycle[1:]
    return cycle
```

```python
# sample dataset
filename = "/content/sample_data/input.txt"
with open(filename) as f:
  input = f.readlines()
  input = [x.strip() for x in input]

graph = input_to_graph(input)
```

```python
initial_node = random.choice(list(graph.keys()))
eulerian_cycle = find_eulerian_cycle(graph,initial_node)
```

```python
if eulerian_cycle[0] == eulerian_cycle[-1]:
  print('->'.join(eulerian_cycle))
else:
  print('Nema Eulerovog ciklusa.')
```
