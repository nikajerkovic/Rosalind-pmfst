## BA3A

**String Composition Problem**

Generate the k-mer composition of a string.

Given: An integer k and a string Text.

Return: Compositionk(Text) (the k-mers can be provided in any order).

Link: https://rosalind.info/problems/ba3a/

```python
def Composition(k, text):
  composition = []
  for i in range(len(text)-k+1):
    composition.append(text[i:i+k])

  # lexicographic order (iako prihvaca rjesenje i bez da je sortirano)
  sorted_composition = sorted(composition)
  return sorted_composition
```
