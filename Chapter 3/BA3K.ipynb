## BA3K

**Contig Generation Problem**

Generate the contigs from a collection of reads (with imperfect coverage).

Given: A collection of k-mers Patterns.

Return: All contigs in DeBruijn(Patterns). (You may return the strings in any order.

Link: https://rosalind.info/problems/ba3k/

Funkcija iz prethodnog zadatka

```python
# BA3E
def  DeBruijnGraph_from_kmers(Patterns):

  nodes = {}

  for k_mer in Patterns:
    if k_mer[:-1] not in nodes.keys():
      nodes[k_mer[:-1]] = list()
      nodes[k_mer[:-1]].append(k_mer[1:])
    else:
      nodes[k_mer[:-1]].append(k_mer[1:])

  return nodes
```

```python
# sample dataset
with open('/content/sample_data/input.txt') as input_data:
    k_mers = [line.strip() for line in input_data.readlines()]
```

```python
# dataset
with open('/content/rosalind_ba3k.txt') as input_data:
    k_mers = [line.strip() for line in input_data.readlines()]
```

```python
edges =DeBruijnGraph_from_kmers(k_mers)
```

```python
out= []
for value in edges.values():
  for node in value:
    out.append(node)
```

```python
balanced= []
unbalanced = []
# set jer reducira ponavljanje 
for node in set(out + list(edges.keys())):
    out_value = out.count(node)
    if node not in edges.keys():
        in_value = 0
    else:
        in_value = len(edges[node])

    if in_value == out_value == 1:
        balanced.append(node)
    else:
        unbalanced.append(node)
```

```python
def get_contigs(start,curr,edges,balanced):
  result = []
  for edge in edges[start]:
    if edge not in balanced:
      result.append(curr+edge[-1])
    else:
      result.append(get_contigs(edge,curr+edge[-1],edges,balanced))

  return result
```

```python
contigs = []
for start in set(unbalanced) & set(edges.keys()):
    contigs.append(get_contigs(start,start,edges,balanced))
```

```python
def flattenNestedList(nestedList):
    flatList = []
    for elem in nestedList:
        # provjerava je li element tipa lista
        if isinstance(elem, list):
            flatList.extend(flattenNestedList(elem))
        else:
            flatList.append(elem)    
    return flatList
```

```python
flat = flattenNestedList(contigs)
```
