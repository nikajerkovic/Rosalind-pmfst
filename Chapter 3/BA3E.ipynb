## BA3E

**De Bruijn Graph from k-mers Problem**

Construct the de Bruijn graph from a collection of k-mers.

Given: A collection of k-mers Patterns.

Return: The de Bruijn graph DeBruijn(Patterns), in the form of an adjacency list.

Link: https://rosalind.info/problems/ba3e/

Funkcija iz prethodnog zadatka:

```python
def print_DeBruijn(nodes):
  graph = [' -> '.join([node[0], ','.join(sorted(node[1]))]) for node in sorted(nodes.items())]
  print('\n'.join(graph))
```

Novo:

```python
def  DeBruijnGraph_from_kmers(Patterns):
  
  nodes = {}

  for k_mer in Patterns:

    if k_mer[:-1] not in nodes.keys():
      nodes[k_mer[:-1]] = list()
      nodes[k_mer[:-1]].append(k_mer[1:])
    else:
      nodes[k_mer[:-1]].append(k_mer[1:])

  return nodes
```

```python
# sample dataset
nodes = DeBruijnGraph_from_kmers(["GAGG","CAGG","GGGG","GGGA","CAGG","AGGG","GGAG"])
print_DeBruijn(nodes)
```

```python
# dataset 
with open('/content/rosalind_ba3e.txt') as input_data:
    Patterns = [line.strip() for line in input_data.readlines()]
```

```python
nodes = DeBruijnGraph_from_kmers(Patterns)
print_DeBruijn(nodes)
```
