## BA3C

**Overlap Graph Problem**

Construct the overlap graph of a collection of k-mers.

Given: A collection Patterns of k-mers.

Return: The overlap graph Overlap(Patterns), in the form of an adjacency list.

Link: https://rosalind.info/problems/ba3c/

```python
def Overlap(sequence):
  result_list = []
  for i in range(len(sequence)):
    current_string = sequence[i]
    for element in sequence:
      if current_string[1:] == element[:-1] and current_string != element:
        result_list.append(current_string+' -> '+element)

  return result_list
```

```python
# sample dataset
result_list = Overlap(["ATGCG","GCATG","CATGC","AGGCA","GGCAT"])
```

```python
for element in result_list:
  print(element)
```

```python
# dataset
with open('rosalind_ba3c (1).txt') as f:
    lines = [line.rstrip('\n') for line in f]
```

```python
result_list = Overlap(lines)
```

```python
for element in result_list:
  print(element)
```
