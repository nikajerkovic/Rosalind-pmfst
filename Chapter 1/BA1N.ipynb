## BA1N

The d-neighborhood Neighbors(Pattern, d) is the set of all k-mers whose Hamming distance from Pattern does not exceed d.

**Generate the d-Neighborhood of a String**

Find all the neighbors of a pattern.

Given: A DNA string Pattern and an integer d.

Return: The collection of strings Neighbors(Pattern, d).

Link: https://rosalind.info/problems/ba1n/

```python
# BA1G
def hamming(dna1,dna2):
  counter=0
  for i, j in zip(dna1,dna2):
    if i != j:
      counter+=1

  return counter
```

```python
# BA1I
def make_kmers(k):
  a_string = 'ACTG'
  final_list = [[]]
  length = k
  groups = [list(a_string)] * length
  for i in groups:
      final_list = [x+[y] for x in final_list for y in i]
  permutations = [''.join(item) for item in final_list]
  return permutations
```

Novo:

```python
def GenerateNeighborhood(pattern,d):
  neighborhood = []
  k = len(pattern)
  k_mers = make_kmers(k)

  for k_mer in k_mers:
    if hamming(pattern,k_mer) <= d:
      neighborhood.append(k_mer)

  return neighborhood
```

```python
# sample dataset
neighborhood = GenerateNeighborhood('ACG',1)
print(*neighborhood, sep = "\n")
```

```python
# dataset
neighborhood = GenerateNeighborhood('ATGAATCGT',2)
print(*neighborhood, sep = "\n")
```
