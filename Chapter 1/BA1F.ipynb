```python
import numpy as np
```

```python
def Skew(sequence):
    length = len(sequence)
    skew = 0
    arr = []
    arr.append(skew)
    for i in range(length):
        if sequence[i]=='G':
            skew = skew + 1
        elif sequence[i]=='C':
            skew = skew -1
        else:
            skew = skew + 0
        arr.append(skew)
    np_arr = np.array(arr)
    return np_arr
```

```python
def find_min(arr):
    min = []
    for i in range(len(arr)):
        if arr[i]==np.min(arr):
            min.append(i)
    np_min = np.array(min)
    return np_min
```

```python
# sample dataset
result = Skew('CCTATCGGTGGATTAGCATGTCCCTGTACGTTTCGCCGCGAACTAGTTCACACGGCTTGATGGCAAATGGTTTTTCCGGCGACCGTAATCGTCCACCGAG')
```

```python
minimum = find_min(result)
```

```python
print(' '.join(map(str, minimum))) 
```

```python
result = Skew('')
minimum = find_min(result)
```

```python
print(' '.join(map(str, minimum))) 
```
