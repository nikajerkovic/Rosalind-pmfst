## BA1K

Given an integer k, we define the frequency array of a string Text as an array of length 4k, where the i-th element of the array holds the number of times that the i-th k-mer (in the lexicographic order) appears in Text (see Figure 1.

**Computing a Frequency Array**

Generate the frequency array of a DNA string.

Given: A DNA string Text and an integer k.

Return: The frequency array of k-mers in Text.

Link: https://rosalind.info/problems/ba1k/

```python
# BA1I
def make_kmers(k):
  a_string = 'ACTG'
  final_list = [[]]
  length = k
  groups = [list(a_string)] * length
  for i in groups:
      final_list = [x+[y] for x in final_list for y in i]
  permutations = [''.join(item) for item in final_list]
  return permutations
```

Novo:

```python
def ComputingFrequencies(text,k):
  k_mers = sorted(make_kmers(k))

  freq_dict = {}

  for k_mer in k_mers:
    counter = 0
    for i in range(len(text)-k+1):
      if k_mer == text[i:i+k]:
        counter+=1
    
    freq_dict[k_mer] = counter

  return list(freq_dict.values())
```

```python
# sample dataset
result = ComputingFrequencies('ACGCGGCTCTGAAA',2)
```
