## BA4J

**Linear Spectrum Problem**

Generate the ideal linear spectrum of a peptide.

Given: An amino acid string Peptide.

Return: The linear spectrum of Peptide.

Link: https://rosalind.info/problems/ba4j/

```python
def get_mass():
    mass = {
        "G": 57,
        "A": 71,
        "S": 87,
        "P": 97,
        "V": 99,
        "T": 101,
        "C": 103,
        "I": 113,
        "L": 113,
        "N": 114,
        "D": 115,
        "K": 128,
        "Q": 128,
        "E": 129,
        "M": 131,
        "H": 137,
        "F": 147,
        "R": 156,
        "Y": 163,
        "W": 186,
    }

    return mass
```

```python
def LinearSpectrum(peptide):

  mass = get_mass()
  prefix_mass  = [0]
  linear_spectrum = [0]
 
  for i in range(len(peptide)):
    current = prefix_mass[i] + mass[peptide[i]]
    prefix_mass.append(current)

  for i in range(len(peptide)):
    for j in range(i + 1, len(peptide) + 1):
      linear_spectrum.append(prefix_mass[j] - prefix_mass[i])
  
  return sorted(linear_spectrum) 
```

```python
# sample dataset
LinearSpectrum('NQEL')
```

```python
# dataset
result = LinearSpectrum('DICMWDQFWLYNGKFHDTFFWINDDQQRSLKQCSWYTTPLEHSDMWTSK')
```
