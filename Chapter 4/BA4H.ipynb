## BA4H

**Spectral Convolution Problem**

Compute the convolution of a spectrum.

Given: A collection of integers Spectrum.

Return: The list of elements in the convolution of Spectrum in decreasing order of their multiplicities. If an element has multiplicity k, it should appear exactly k times.

Link: https://rosalind.info/problems/ba4h/

```python
from collections import Counter
```

```python
def ConvolutionOfSpectrum(spectrum):

  difference = []

  for i in range(1,len(spectrum)):
    for j in range(i):
      if spectrum[i] != spectrum[j]:
        difference.append(abs(spectrum[i]-spectrum[j]))

  # Counter(difference).most_common() vraca parove: vrijednost - koliko puta se pojavljuje
  return [item for items, c in Counter(difference).most_common() for item in [items] * c]
```

```python
# sample dataset
ConvolutionOfSpectrum([0,137, 186, 323])
```

```python
with open("/content/rosalind_ba4h.txt") as f:
  spectrum = [int(s) for s in list(f.read().strip().split())]
```

```python
result = ConvolutionOfSpectrum(spectrum)
```
