## BA4B

**Peptide Encoding Problem**

Find substrings of a genome encoding a given amino acid sequence.

Given: A DNA string Text and an amino acid string Peptide.

Return: All substrings of Text encoding Peptide (if any such substrings exist).

Link: https://rosalind.info/problems/ba4b/

Funkcije iz prethodnog zadatka:

```python
def make_codons(string):
  mers = []
  for i in range(0,(len(string)),3):
    mers.append(string[i:i+3])

  return mers
```

```python
def genetic_code_dict():
  dictionary = {
        "AAA": "K",
        "AAC": "N",
        "AAG": "K",
        "AAU": "N",
        "ACA": "T",
        "ACC": "T",
        "ACG": "T",
        "ACU": "T",
        "AGA": "R",
        "AGC": "S",
        "AGG": "R",
        "AGU": "S",
        "AUA": "I",
        "AUC": "I",
        "AUG": "M",
        "AUU": "I",
        "CAA": "Q",
        "CAC": "H",
        "CAG": "Q",
        "CAU": "H",
        "CCA": "P",
        "CCC": "P",
        "CCG": "P",
        "CCU": "P",
        "CGA": "R",
        "CGC": "R",
        "CGG": "R",
        "CGU": "R",
        "CUA": "L",
        "CUC": "L",
        "CUG": "L",
        "CUU": "L",
        "GAA": "E",
        "GAC": "D",
        "GAG": "E",
        "GAU": "D",
        "GCA": "A",
        "GCC": "A",
        "GCG": "A",
        "GCU": "A",
        "GGA": "G",
        "GGC": "G",
        "GGG": "G",
        "GGU": "G",
        "GUA": "V",
        "GUC": "V",
        "GUG": "V",
        "GUU": "V",
        "UAA":"STOP",
        "UAC": "Y",
        "UAG":"STOP",
        "UAU": "Y",
        "UCA": "S",
        "UCC": "S",
        "UCG": "S",
        "UCU": "S",
        "UGA": "STOP",
        "UGC": "C",
        "UGG": "W",
        "UGU": "C",
        "UUA": "L",
        "UUC": "F",
        "UUG": "L",
        "UUU": "F",
    }

  return dictionary
```

```python
#BA4A
def ProteinTranslation(string):
  amino_list = []
  codons = make_codons(string)
  dictinary = genetic_code_dict()

  for mer in codons:
    if dictinary[mer] ==  'STOP':
      break
    else:
      amino_list.append(dictinary[mer])
  
  return amino_list
```

```python
# BA1C
def reverseComplement(pattern):
  r_complement = ''
  r_dict = {
      "A": "T","C":"G","G":"C","T":"A"
  }
  for i in pattern[::-1]:
    r_complement += r_dict[i]

  return r_complement
```

Novo:

```python
# funckija za pretvaranje dna u rna i rna u dna
# choice - 1 -> dna to rna
#        - 2 -> rna to dna
def dna_and_rna(pattern,choice):

  if choice == 1:
    pattern = pattern.replace('T','U')
  elif choice == 2:
    pattern = pattern.replace('U','T')

  return pattern
```

```python
def PeptideEncoding(dna, peptide):
  
    substrings = []
    rna = dna_and_rna(dna,1)
    reverse_rna = dna_and_rna(reverseComplement(dna),1)
    sub_len = len(peptide)*3
    start = 0


    while start + sub_len <= len(dna):
      sub_rna = rna[start: start + sub_len]
      sub_rev_rna = reverse_rna[start: start + sub_len]

      if peptide == ProteinTranslation(sub_rna):
        substrings.append(dna_and_rna(sub_rna,2))

      if peptide == ProteinTranslation(sub_rev_rna):
    
        substrings.append(reverseComplement(dna_and_rna(sub_rev_rna,2)))
        

      start += 1

    return substrings
```
